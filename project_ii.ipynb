```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# show all columns on pandas dataframes
pd.set_option('display.max_columns', None)
%reload_ext rpy2.ipython
```

```python
df = pd.read_csv('/Users/mac/Library/Containers/com.microsoft.Excel/Data/Downloads/visa.csv')
df = df.dropna(subset=['Country'])
df = df.dropna(subset=['Total'])
df['Country'] = df['Country'].str.strip()
df.drop(['Year'], axis=1, inplace=True)
df
```

```python
#df = df.apply(lambda x: x.str.replace(',',''))

df['Total'] = df['Total'].str.replace(',','')
df
```

```python
df.Total = pd.to_numeric(df.Total, errors='coerce')
df
```

```python
df.Total = pd.to_numeric(df.Total, errors='coerce')
df
```

```python
grouped_category_year = df.groupby(['Country']).agg(np.sum).reset_index()
grouped_category_year
```

```python
grouped_category_year.sort_values(by='Total', ascending=False, inplace=True)
grouped_category_year
```

The next line shows the top ten countries US issued student visa approvals in the last ten years.

```python
#grouped_category_year.head()
grouped_category_year.head(10)
```

```python
grouped_category_year.head(10).to_csv('visa_approvals.csv')
```

```python
# Load the data into a pandas DataFrame
df = pd.read_csv("visa_approvals.csv")
df.plot(kind='bar', x='Country', y='Total', color='#00ba38')
                                                                                                                                                                                                                                                        
# Plot the bar chart
plt.bar(x='lab', y='val', rot=1)


# Add a title to the chart
plt.title("Top ten countries with the highest US immigrant student visa approvals in ten years")
plt.xlabel("Country")
plt.ylabel("Total Number of Approvals")

# Show the plot
plt.show(10)
```
